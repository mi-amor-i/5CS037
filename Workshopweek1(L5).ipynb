```python
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])
A_inv = np.linalg.inv(A)
print(np.dot(A,A_inv))
```

```python
AB= np.dot(A,B)
BA= np.dot(B,A)
print(np.array_equal(AB,BA))
```

```python
AB_T = AB.T
BT_AT = np.dot(B.T,A.T)
print(np.array_equal(AB_T,BT_AT))
```
