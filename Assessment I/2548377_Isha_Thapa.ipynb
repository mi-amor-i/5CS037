3. Outlier Detection:  
• Detect outliers in HDI and GNI per Capita using the 1.5 × IQR rule.  
• Create a scatter plot of GNI per Capita vs HDI, highlighting the outliers in a different color.   
• Discuss why the identified countries stand out as outliers.


```python
hdi_Q1 = hdi_south_asian_df['hdi'].quantile(0.25)
hdi_Q3 = hdi_south_asian_df['hdi'].quantile(0.75)
hdi_IQR = hdi_Q3 - hdi_Q1

lower_bound_hdi = hdi_Q1 - 1.5 * hdi_IQR
upper_bound_hdi = hdi_Q3 + 1.5 * hdi_IQR

hdi_outliers = hdi_south_asian_df[(hdi_south_asian_df['hdi'] < (lower_bound_hdi)) | (hdi_south_asian_df['hdi'] > upper_bound_hdi)]

gni_Q1 = hdi_south_asian_df['gross_inc_percap'].quantile(0.25)
gni_Q3 = hdi_south_asian_df['gross_inc_percap'].quantile(0.75)
gni_IQR = gni_Q3 - gni_Q1

lower_bound_gni = gni_Q1 - 1.5 * gni_IQR
upper_bound_gni = gni_Q3 + 1.5 * gni_IQR

gni_outliers = hdi_south_asian_df[(hdi_south_asian_df['gross_inc_percap'] < lower_bound_gni) | (hdi_south_asian_df['gross_inc_percap'] > upper_bound_gni)]

print("Outliers in HDI:")
print("HDI Outliers:",hdi_outliers)

print("\nOutliers in GNI per Capita:")
print("GNI Outliers:",gni_outliers)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(hdi_south_asian_df['gross_inc_percap'], hdi_south_asian_df['hdi'], color='blue', label='Normal Countries')

if len(gni_outliers) > 0:
    plt.scatter(gni_outliers['gross_inc_percap'], gni_outliers['hdi'], color='red', s=100, label='Outliers')

plt.xlabel('GNI per Capita')
plt.ylabel('HDI')
plt.title('GNI per Capita vs HDI (Outliers in Red)')
plt.legend()
plt.grid(True, alpha=0.3)

plt.show()
```

Outliers is caused in countries because the values in dataset don't follow general pattern like others. Countries with high GNI outliers means the country is economically prosperous whereas low GNI outliers means the country is economically limited country.

7. Outlier Detection:  
• Detect outliers in HDI {"hdi"} and GNI per Capita {"gross inc percap"} for each region using the 1.5 × IQR rule.  
• Create scatter plots highlighting outliers in a different color.  
• Discuss the significance of these outliers.


```python
def outliers_calculation (df, column):
  Q1 = df[column].quantile(0.25)
  Q3 = df[column].quantile(0.75)
  IQR = Q3 - Q1

  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 + IQR

  return (df[column] < lower_bound) | (df[column] > upper_bound)

sa_hdi_outliers = outliers_calculation(south_asian_df, 'hdi')
sa_gni_outliers = outliers_calculation(south_asian_df, 'gross_inc_percap')

me_hdi_outliers = outliers_calculation(middle_east_df, 'hdi')
me_gni_outliers = outliers_calculation(middle_east_df, 'gross_inc_percap')

```

```python
plt.figure(figsize=(10, 6))
is_any_outlier  = sa_hdi_outliers | sa_gni_outliers

plt.scatter(south_asian_df[~is_any_outlier ]['gross_inc_percap'], south_asian_df[~is_any_outlier ]['hdi'], color='blue', label='Normal Countries', alpha=0.7)
plt.scatter(south_asian_df[is_any_outlier ]['gross_inc_percap'], south_asian_df[is_any_outlier ]['hdi'], color='red', s=100, label='Outliers', edgecolor='black')

plt.xlabel('GNI per Capita')
plt.ylabel('HDI')
plt.title('GNI per Capita vs HDI (Outliers in Red) South Asian')
plt.legend()
plt.grid(True, alpha=0.3)

plt.show()
```

```python
plt.figure(figsize=(10, 6))
is_any_outlier  = me_hdi_outliers | me_gni_outliers

plt.scatter(middle_east_df[~is_any_outlier ]['gross_inc_percap'], middle_east_df[~is_any_outlier ]['hdi'], color='blue', label='Normal Countries', alpha=0.7)
plt.scatter(middle_east_df[is_any_outlier ]['gross_inc_percap'], middle_east_df[is_any_outlier ]['hdi'], color='red', s=100, label='Outliers', edgecolor='black')

plt.xlabel('GNI per Capita')
plt.ylabel('HDI')
plt.title('GNI per Capita vs HDI (Outliers in Red) Middle East')
plt.legend()
plt.grid(True, alpha=0.3)

plt.show()
```
